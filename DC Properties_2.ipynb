```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.read_csv('DC_Properties.csv')
```

```python
#Distribution of when Residential Properties were built
x = df1['AYB']
plt.hist(x, bins=20)
plt.xlabel('Year Built')
plt.show()
```

Based on the data from the 2010 Census, the majority of DC properties were built between 1900 and 1950. While we don't have data after 2010, it does appear that starting in 2000, more properties were being built in increasing amounts. It will be interesting to see in the next Census if that upward trend continued. 

```python
#Distribution of when Remodels were Completed
x = df1['YR_RMDL']
plt.hist(x, bins=100, color='green')
plt.xlim(1900, 2010)
plt.xlabel('Year Remodeled')
plt.show()
```

In the previous graph, we were able to see that the majority of properties were built before 1950. One could assume then that a large amount of remodels would start to occur at some point after 1950. The data shows that starting in 1960, the number of remodels has increased year after year, and quite substantially in 2000. 

```python
#Grouping Buildings built before 1950
pre_1950 = pd.DataFrame(df1.loc[lambda df: df1['AYB'] < 1950, :])

#Grouping Buildings built after 1950
post_1950 = pd.DataFrame(df1.loc[lambda df: df1['AYB'] > 1950, :])
```

```python
#Comparing Prices of Properties Built Before and After 1950

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(pre_1950['PRICE'])
plt.ylabel('Price')
plt.title('Price of Pre-1950 Properties')

plt.subplot(1, 2, 2)
plt.plot(post_1950['PRICE'], color='purple')
plt.ylabel('Price')
plt.title('Price of Post-1950 Properties')
plt.show()
```

When comparing properties built pre- and post-1950, those built pre-1950 seem to have a higher price on average. It would be interesting to for further research to explore how many of the pre-1950s properties have been remodeled, and of those remodels, what the price is.
